```python
path_1="C:\\Users\\USER PC\\OneDrive\\Desktop\\Projects\\phonepe\\pulse-master\\data\\aggregated\\insurance\\country\\india\\state\\"
```

```python
aggre_insur_list = os.listdir(path_1)

column1={"States":[], "Years":[], "Quarter":[], "Transaction_Type":[], "Transaction_Count":[], "Transaction_Amount":[]}
for state in aggre_insur_list:
    current_states=path_1+state+"\\"
    agg_year_list=os.listdir(current_states)
    
    for year in agg_year_list:
        current_year=current_states+year+"\\"
        agg_year_data=os.listdir(current_year)
        
        
        for file in agg_year_data:
            current_data=current_year+file
            data=open(current_data,"r")
            
            
            A=json.load(data)
            
            
            for i in A["data"]["transactionData"]:
                name=i["name"]
                insurance_count=i["paymentInstruments"][0]["count"]
                insurance_amount=i['paymentInstruments'][0]['amount']
                column1["Transaction_Type"].append(name)
                column1["Transaction_Count"].append(insurance_count)
                column1["Transaction_Amount"].append(insurance_amount)
                column1["States"].append(state)
                column1["Years"].append(year)
                column1["Quarter"].append(int(file.strip(".json")))
                
            
```

```python
aggre_insurance = pd.DataFrame(column1)

aggre_insurance["States"]=aggre_insurance["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar Island')
aggre_insurance["States"]=aggre_insurance["States"].str.replace('-',' ')
aggre_insurance["States"]=aggre_insurance["States"].str.title()
aggre_insurance["States"]=aggre_insurance["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu ')
```
